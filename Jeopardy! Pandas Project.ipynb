```python
#some column names have space in front of the names, so we need to delete the space
jeopardy = jeopardy.rename(columns = {
    ' Air Date': 'Air Date',
    ' Round': 'Round',
    ' Category': 'Category',
    ' Value': 'Value',
    ' Question': 'Question',
    ' Answer': 'Answer'
})
jeopardy.head()
```

```python
#a function that filters the dataset for questions that contains all of the words in a list of words
def num_of_question(words):
    filter1 = lambda x: all(word.lower() in x.lower() for word in words)
    return jeopardy[jeopardy['Question'].apply(filter1)]
#     return jeopardy.loc[jeopardy['Question'].apply(lambda_function)]

print(num_of_question(['King', 'England']))
```

```python
#a function that only matches the exact word in the rows
import re

def num_of_question1(words):
    filter2 = lambda x: all(re.search(r'\b' + word.lower() + r'\b', x.lower()) for word in words)
#     return jeopardy[jeopardy['Question'].apply(filter2)]
    return jeopardy.loc[jeopardy['Question'].apply(filter2)]

print(num_of_question1(['art', 'show']))
```

```python
#change the datatype of column Value from string to float
import numpy as np

## convert value None into 0 for column Value
ignore_none = lambda value: '0' if value == 'None' else value

jeopardy['Value'] = jeopardy['Value'].replace(['[\$]', '[\,]'], '', regex=True).apply(ignore_none)

jeopardy['Value'] = pd.to_numeric(jeopardy['Value'])

jeopardy['Value']
```

```python
jeopardy.dtypes
```

```python
#a function to calculate average value to questions including the words in the list
def average_value_to_question(words):
    #select rows with words included in the questions
    selection = num_of_question1(words)
    #exclude rows with value 0
    selection = selection[selection['Value'] > 0]
    return selection['Value'].mean()

print(average_value_to_question(['art', 'show']))
```

```python
#a function that returns the count of unique answers to the questions including the words in the list
def count_unique_answers(words):
    selection = num_of_question1(words)
    selection = selection[selection['Value'] > 0]
    return selection['Answer'].nunique()

print(count_unique_answers(['art', 'show']))
```

```python
#find any relations bewteen Round and Category
category_round = jeopardy.groupby(['Round', 'Category'])['Question'].count().reset_index()
category_round
```

```python
#which category has the most questions in round Double Jeopardy?
double_jeopardy = category_round[category_round['Round'] == 'Double Jeopardy!']
double_jeopardy.sort_values(by=['Question'], ascending=False)
```

```python
#which category has the most questions in round Jeopardy?
jeopardy_round = category_round[category_round['Round'] == 'Jeopardy!']
jeopardy_round.sort_values(by=['Question'], ascending=False)
```

```python
#which category has the most questions in round Final Jeopardy?
final_jeopardy = category_round[category_round['Round'] == 'Final Jeopardy!']
final_jeopardy.sort_values(by=['Question'], ascending=False)
```

```python
#which category has the most questions in round Tiebreaker?
tiebreaker = category_round[category_round['Round'] == 'Tiebreaker']
tiebreaker.sort_values(by=['Question'], ascending=False)
```

```python
#a functino that finds the dates where the question that contains specific words appears
def dates(words):
    selection = num_of_question1(words)
    selection = selection[selection['Value'] > 0]
    
    ##use groupby function to see how many questions that contain the words are in each date
    dates_num_of_question = selection.groupby(['Air Date'])['Question'].count().reset_index()
    
    ##separate dates into before 2000 and after 2000
    before_2000 = dates_num_of_question[dates_num_of_question['Air Date'] < '2000-1-1']
    after_2000 = dates_num_of_question[dates_num_of_question['Air Date'] >= '2000-1-1']
    
    ##create a dictionary that stores how many questions containing the word are before or after 2000
    date_dict = {}
    date_dict['Before 2000'] = len(before_2000)
    date_dict['After 2000'] = len(after_2000)
    
    return date_dict
    
print(dates(['computer']))
print(dates(['movie']))
```
